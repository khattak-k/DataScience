```python
#import libraries
import folium
import pandas as pd
```

```python
#create a map on Pakistan location
pakistan_map = folium.Map(location = [30.37,69.34],zoom_start=5,tiles='OpenStreetMap')
pakistan_map
```

```python
#load the dataset (csv file) into the Pandas dataframe
df = pd.read_csv('PK COVID-19-3jun.csv')
df.head()
```

```python
#calcute total cases in each province
total_cases = df['Cases'].sum()
cases = df.groupby('Province').sum()
provinces = df['Province'].unique().tolist()
coordinates = [[33.7205, 73.0405],[25.8943, 68.5247],[35.8026, 74.9832],[28.4907, 65.0958],[31.1704, 72.7097],[34.9526, 72.3311],[33.9259, 73.7810],[32.6675, 69.8597]]
#make dict of provinces and coordinates
prov_coord = dict(zip(provinces,coordinates))
#add coordinates in df as a new column if province matches with df['Province']
df['Coordinates'] = df['Province'].map(prov_coord)
cases = cases.reset_index()
cases,


```

```python
#mark circles on the map and displa the map
for state,tc,de,re in zip(list(cases['Province']),list(cases['Cases']),list(cases['Deaths']),list(cases['Recovered'])):
    folium.CircleMarker(location = prov_coord[state],
                       radius = (tc/total_cases)*150,
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>State:  ' + state + '</strong><br>' + 'Total Cases: ' + str(tc) + '<br>' + 'Deaths: '+ str(de) + '<br>' + 'Recoveries: ' + str(re)
    iframe = folium.IFrame(xyz, width=200, height=150)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = prov_coord[state], popup=popupData).add_to(pakistan_map)
    

pakistan_map
```
